#### ***Data Description***

The dataset contains **12 columns** which records the various attributes of the passengers aboard the titanic. The data dictionary is given below
- PassengerID - unique identifier
- survived - If passenger will survive or not (Yes - 1, No - 0)
- pclass   - Ticket class (1st, 2nd, and 3rd)
- sex      - Sex of the passenger (M / F)
- Age      - Age in years
- sibsp    - The number of siblings / spouses aboard the Titanic
- parch    - The number of parents / children aboard the Titanic
- ticket   - Ticket number
- fare     - Passenger fare
- cabin    - Cabin number
- embarked - Port of embarkation

#### ***Summary***

<b><i>EDA & Data Preprocessing </i></b>

The training dataset provided had 891 observations and 12 features. <br>

Null values were observed in the Age, Embarked and Cabin columns.<br>
Age  - 19.8% null values<br> 
Cabin - 77 % null values <br>
Embarked - 0.002 % null values <br>

In the dataset, approximately 15% of the survived passengers had missing age values. To address this, an imputation method was employed by calculating the median age for ***each combination of passenger class (Pclass) and gender***. This approach aimed to estimate the missing ages more accurately by taking into account the specific characteristics of each subgroup and yielded significantly different results compared to simply imputing the median age of the entire passenger population. <br>

The analysis also involved identifying children among the passengers, specifically those with the title 'Master' as this title was associated with boys under 18. For the missing age values within this subgroup, ***the median age of the children under 18*** was used for imputation. .The ***subgroup-specific imputations*** based on Pclass and gender, as well as accounting for children under 12, the imputation process was aimed to enhance the accuracy of age estimation and contribute to improved model performance.<br>

The cabin column was dropped after extracting meaning information about the number of cabins allocate to each family. The ,issing values of Embarked was replaced with the 'C' after analysing the fare and cabin values associated the missing location an dcomparing it with data from embarkment location.
No duplicates were found.<br>

Outliers were detected in the fare and age columns.<br>

***Feature engineering*** was performed to enhance the dataset by creating new informative features. <br>
- Title Extraction: The passengers' titles were extracted from the name column. This process likely involved parsing the names to identify titles such as "Mr.", "Mrs.", "Miss", etc. These provide insights into social status or age groupings and the passengers who might have ben given priority for life boat usage.

- Individual Fare Calculation: To obtain a more granular measure of fare per passenger, the fare column was divided by the family size. This calculation aimed to capture the fare allocation on an individual basis, accounting for variations due to family size.

- Family Size Determination: Family size was computed by adding the values from the SibSp (number of siblings/spouses aboard) and Parch (number of parents/children aboard) columns.

- Cabin Analysis: Despite a majority of missing cabin values, certain information could still be extracted. It was observed that passengers traveling with siblings/spouses and children tended to have higher fares. The cabin column, which listed all the cabin numbers allocated to the entire family, was examined to identify the decks assigned to each family and the number of cabins occupied. However, due to the substantial number of missing cabin values, the decision was made to extract only the number of cabins from the cabin column. Missing cabin values were replaced with 0 to indicate the absence of cabin information.

The ticket, passenger ID, and name columns were dropped. These columns contained unique identifiers or redundant information.<br>

The analysis revealed several notable observations regarding the variables' relationship with the target variable.<br>

Firstly, a majority of the survivors were found to be female and belonged to the 1st class, indicating a potential correlation between gender, passenger class, and survival rates.<br>

A significant number of passengers embarked from the 'S' location and experienced higher casualties. This finding can be partly attributed to the fact that the majority of passengers embarking from 'S' were males belonging to the 3rd class, suggesting a potential correlation between embarkation location, gender, passenger class, and survival outcomes.<br>

Upon examining the correlation matrix it was observed that the passenger class (Pclass) exhibited a ***weak negative correlation*** with the target variable, suggesting that a higher class may have increased the chances of survival. On the other hand, variables such as ***fares and cabin_nums showed a weak positive correlation*** with the target variable, indicating a potential association between higher fares, more cabin numbers, and increased survival rates.<br>

These observations highlight the importance of gender, passenger class, embarkation location, fares, and the number of cabins in understanding the factors influencing survival on the Titanic. The correlation matrix provided valuable insights into the strength and direction of these relationships, aiding in further analysis and decision-making processes.<br>

The features selected were Title, Ind fare, family size, Age, Pclass, Sex,Cabin_nums. Parch, SibSp were dropped as after familysize feature development, these features were redundant.

***Model training and Evaluation*** : <br>

A variety of classification models were trained to address the classification problem at hand. 
The following models were employed:
- Decision tree
- Random forest, 
- Adaboost, 
- Logistic Regression, 
- SVC
- Gradient Boosting Classifier
- Voting classifier

The class distribution indicated a class imbalance, with only 39% of the dataset representing survived passengers.To ensure a balanced representation of the classes in the training and testing datasets, the dataset was split using the ***stratify*** option. Models were evaluated with sample generated by both stratified  and non stratified sampling. The models exhibited better testing scores using the non stratified sampling technique.<br>

The categorical variables were converted to indicator variables. For the SVC and logistic regression models a pipeline was created with standardscaler, that scalied the data before fiting the data to the classifiers.<br>

Each of the base models underwent ***hyperparameter tuning*** and ***class weights were assigned***  to optimize their performance. The objective was to find the best set of hyperparameters that would yield improved results.<br>

Among the models, the ***hyperparameter-tuned Random Forest Classifier*** & ***Hyperparameter tuned Logistic Regression***exhibited the most promising performance during validation.The logistic Regresssion model was a more balanced model with 0.82 precision and 0.84 recall but with a slightly lower auc score of 0.90. An attempt was made to develop a better model by training a Voting Classifier using the best performing models. However, the improvement in the metrics achieved by the ensemble model was not significant enough to justify its use over the hyperparameter-tuned Random Forest model.<br>

Considering the overall performance and evaluation results, the recommendation is to employ the hyperparameter-tuned Random Forest model for this classification problem. ##### ***Null Values And Imputation Of Values***


```python
#Checking for null values
a = [np.nan, None, [], {}, 'NaN', 'Null','NULL','None','NA','?','-', '.','', ' ', '   ']
for c in data.columns:
    string_null = np.array([x in a for x in data[c]])
    print(c, data[c].isnull().sum(), string_null.sum()) 
```

```python
#Calculating the % of survivors with null age values
sa = train_data[train_data['Age'].isnull()]
len(sa[sa['Survived']==1])/len(train_data[train_data['Survived']==1])
```

```python
#Calculating the % of null values in each column
train_data.isnull().sum()/len(train_data)
```

```python
# Finding the median age of the each Pclass + Sex subgroup
grouped_median_age = data.groupby(['Sex', 'Pclass'])['Age'].median()
print(grouped_median_age)

#Function to impute age based on the gender and Pclass
def impute_age(row):
    if pd.isnull(row['Age']):
        return grouped_median_age[row['Sex'], row['Pclass']]
    return row['Age']

data['Age'] = data.apply(impute_age, axis=1) # replacing nulls with median age based on the gender & Pclass

data['Cabin'] = data['Cabin'].fillna('No') #replacing nulls in cabin with No

data['Embarked'] = data['Embarked'].fillna('C') # replacing the nulls in embarked with the mode value
```

##### ***Feature Engineering***


```python
#Generating the Title column

data['Title'] = train_data['Name'].str.split(', ').str[1].str.split('.').str[0].str.strip() # generating the title from the names provided
print(data['Title'].unique())
title = ['Don', 'Rev','Dr' ,'Mme', 'Ms' ,'Major' ,'Lady','Sir', 'Mlle' ,'Col', 'Capt', 'the Countess','Jonkheer']
replacement = ['Other']*len(title)

data['Title'].replace(['Ms','Mlle'],['Miss','Miss'],inplace = True)
data['Title'].replace(['Ms','Mme'],['Mrs','Mrs'],inplace = True)
data['Title'].replace(title,replacement,inplace = True) # Replacing all the titles other than Mr,Mrs, Miss, Master with others

print(data['Title'].unique())
```

```python
#Creating a new column to add family size
data['Familysize'] = data['SibSp'] + data['Parch'] + 1
```

```python
# Creating # of cabins for each family based on the cabin column anf family size
cabin_num =[]
cabin_loc = []
for i in data['Cabin']:
    if i =='No':
        cabin_num.append(0)
        cabin_loc.append('XX')
    else:
        cabin_num.append(len(i.split()))
        cabin_loc.append(i[0])

data['cabin_nums'] = cabin_num
data['cabin_loc'] = cabin_loc

#creating new column avgfare price based on the Family size
data['indfare'] = data['Fare'] / data['Familysize']
```

#### ***Data Preparation for Supervised Learning Models***

```python
#creating the target variable
y = data['Survived']

# dropping the traget variable from the train data
X = data.drop(['Survived'], axis =1)

# Feature selection
#Creating an list for the features to be dropped
drop_features = ["Cabin",'Embarked','Fare','cabin_loc','Parch','SibSp']
X.drop(drop_features,axis = 1,inplace = True)

test_features = list(X.columns) # creating a list of columns to be used on filter the test data

X = pd.get_dummies(X) # Coverting the categorical columns 
```

#### ***Splitting the data to train and validation datasets for model training & validation***

```python
#Test train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**Function to print the accuracy score, confusion matrix and classification report**

```python
#creating a function to print the accuracy score, confusionmatrix and classification report

def results(predictdata,targetdata,clf):
    y_pred = clf.predict(predictdata)
    print(f"Accuracy:{round(accuracy_score(targetdata, y_pred),3)}")
    print(f"Precision:{round(precision_score(targetdata, y_pred),3)}")
    print(f"Recall:{round(recall_score(targetdata, y_pred),3)}")
    print("Classification Report :")
    print(classification_report(targetdata, y_pred))
    cm = confusion_matrix(targetdata, y_pred)
    plt.figure(figsize = (3, 2))
    sns.heatmap(cm, annot = True,  fmt = '.2f', xticklabels = [0,1], yticklabels = [0, 1])
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
    
```

**Function to compute the ROC curve and the area under the curve and plot the ROC curve**

```python
from sklearn.metrics import auc, roc_curve
def roc_plot(clf,clf_name):
    plt.figure(figsize=(5, 3))
    y_pred_prob = clf.predict_proba(X_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
    auc_score = auc(fpr, tpr)
    plt.plot(fpr, tpr, label = f"{clf_name} (AUC = {round(auc_score,3)})")
    plt.plot([0, 1], [0, 1], 'k--')  # Plot the diagonal line
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.suptitle(f'{clf_name} - ROC Curve')
    plt.legend(loc='lower right')
    plt.show()
```

```python
dt = DecisionTreeClassifier(random_state =0)
dt.fit(X_train, y_train)
print('Training score ')
results(X_train,y_train,dt)
print('Testing score')
results(X_test,y_test,dt)
roc_plot(dt,'Decision Tree')
```

### <b><i>Hyperparameter Tuned Decision Tree</i></b>
---

```python
param_grid = {
    'max_depth': range(2,10,1),  
    'min_samples_split': range(2,10,1), 
    'min_samples_leaf': range(2,10,1),
    'criterion' :['gini','entropy']
}
grid_search = GridSearchCV(DecisionTreeClassifier(class_weight={0:0.3,1:0.7} ,random_state = 0), param_grid, cv=5,scoring="roc_auc")
grid_search.fit(X_train, y_train)

dt1 = grid_search.best_estimator_
dt1.fit(X_train,y_train)
print('Training score ')
results(X_train,y_train,dt1)
print('Testing score')
results(X_test,y_test,dt1)
roc_plot(dt1,'Decision Tree')
```
